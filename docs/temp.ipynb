```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
data = []

for a in [1, 2, 3]:
    for b in [10, 20, 30]:
        for x in np.arange(10):
            y = a * x ** 2 + b

            data.append([a, b, x, y])

df = pd.DataFrame(data, columns=["a", "b", "x", "y"])
df.sample(10)
```

```python
chart = alt.Chart(df).mark_line().encode(
    x="x:Q",
    y="y:Q",
    color="a:O",
    strokeDash="b:O"
)

chart
```

```python
s1 = alt.binding_range(min=1, max=3, step=1, name='a:')
slider1 = alt.param(value=1, bind=s1)

s2 = alt.binding_range(min=10, max=30, step=10, name='b:')
slider2 = alt.param(value=20, bind=s2)

chart = alt.Chart(df).mark_line().encode(
    x="x:Q",
    y="y:Q",
    detail=["a", "b"],
    color=alt.condition((alt.datum.a == slider1) & (alt.datum.b == slider2), alt.value("red"), alt.value("gray")),
    opacity=alt.condition((alt.datum.a == slider1) & (alt.datum.b == slider2), alt.value(1), alt.value(0.5))
).add_params(slider1, slider2)

chart
```

```python
s1 = alt.binding_range(min=1, max=3, step=1, name='a:')
slider1 = alt.selection_point(name="sa", value=1, fields=["a"], bind=s1)

s2 = alt.binding_range(min=10, max=30, step=10, name='b:')
slider2 = alt.param(name="sb", value=20, bind=s2)

chart = alt.Chart(df).mark_line().encode(
    x="x:Q",
    y="y:Q",
    #detail=["a", "b"],
    #color=alt.condition((alt.datum.a == slider1) & (alt.datum.b == slider2), alt.value("red"), alt.value("gray")),
    #opacity=alt.condition((alt.datum.a == slider1) & (alt.datum.b == slider2), alt.value(1), alt.value(0.5))
).transform_filter(slider1)

chart
```

```python
from vega_datasets import data

cars = data.cars.url

options = ['Europe', 'Japan', 'USA']
labels = [option + ' ' for option in options]

input_dropdown = alt.binding_radio(
    # Add the empty selection which shows all when clicked
    options=options + [None],
    labels=labels + ['All'],
    name='Region: '
)
selection = alt.selection_point(
    fields=['Origin'],
    bind=input_dropdown,
)

alt.Chart(cars).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    # We need to set a constant domain to preserve the colors
    # when only one region is shown at a time
    color=alt.Color('Origin:N').scale(domain=options),
).add_params(
    selection
).transform_filter(
    selection
)
```
